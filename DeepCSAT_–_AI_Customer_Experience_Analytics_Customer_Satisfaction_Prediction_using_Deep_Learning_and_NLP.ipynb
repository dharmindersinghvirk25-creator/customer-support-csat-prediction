## 9 Feature Engineering

```python
# Convert to datetime safely
df['issue_reported_at'] = pd.to_datetime(df['issue_reported_at'], errors='coerce')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], errors='coerce')

# Calculate response time
df['response_time_minutes'] = (
    df['issue_responded'] - df['issue_reported_at']
).dt.total_seconds() / 60
```

## 25.4 Running Streamlit in Google Colab using Ngrok
Since Google Colab does not directly support hosting web applications, Ngrok was used to generate a public URL for accessing the Streamlit application.

Ngrok creates a secure tunnel from the public internet to the local environment where the Streamlit app is running.

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,GRU,Conv1D,Flatten

from docx import Document
from reportlab.pdfgen import canvas

st.set_page_config(page_title="AI Customer Experience Analytics",layout="wide")

st.title("🚀 AI Customer Experience Analytics Dashboard")

# -------------------------------------------------
# LOAD DATA
# -------------------------------------------------

@st.cache_data
def load_data():
    df=pd.read_csv("eCommerce_Customer_support_data.csv")
    df.columns=df.columns.str.lower().str.replace(" ","_")
    return df

df=load_data()

# -------------------------------------------------
# FEATURES
# -------------------------------------------------

features=df[["item_price","connected_handling_time"]].fillna(0)
target=df["csat_score"]

scaler=StandardScaler()
X_scaled=scaler.fit_transform(features)

X=X_scaled.reshape(X_scaled.shape[0],X_scaled.shape[1],1)

X_train,X_test,y_train,y_test=train_test_split(
X,target,test_size=0.2,random_state=42
)

baseline=np.array([[features["item_price"].mean(),
features["connected_handling_time"].mean()]])

baseline=scaler.transform(baseline)
baseline=baseline.reshape(1,2,1)

# -------------------------------------------------
# TRAIN MODELS
# -------------------------------------------------

@st.cache_resource
def train_models():

    models={}
    scores={}
    baseline_csats={}
    histories={}

    def train(model,name):

        history=model.fit(X_train,y_train,epochs=20,batch_size=32,verbose=0)

        preds=model.predict(X_test)

        rmse=np.sqrt(mean_squared_error(y_test,preds))
        r2=r2_score(y_test,preds)

        csat=model.predict(baseline)[0][0]
        csat=float(np.clip(csat,1,5))

        models[name]=model

        scores[name]={"RMSE":rmse,"R2":r2,"CSAT":csat}

        baseline_csats[name]=csat
        histories[name]=history

    dnn=Sequential([
    Flatten(input_shape=(2,1)),
    Dense(64,activation="relu"),
    Dense(32,activation="relu"),
    Dense(1)
    ])
    dnn.compile(optimizer="adam",loss="mse")
    train(dnn,"Deep Neural Network")

    cnn=Sequential([
    Conv1D(32,1,activation="relu",input_shape=(2,1)),
    Flatten(),
    Dense(1)
    ])
    cnn.compile(optimizer="adam",loss="mse")
    train(cnn,"CNN")

    lstm=Sequential([
    LSTM(32,input_shape=(2,1)),
    Dense(1)
    ])
    lstm.compile(optimizer="adam",loss="mse")
    train(lstm,"LSTM")

    gru=Sequential([
    GRU(32,input_shape=(2,1)),
    Dense(1)
    ])
    gru.compile(optimizer="adam",loss="mse")
    train(gru,"GRU")

    return models,scores,baseline_csats,histories

models,MODEL_SCORES,baseline_csats,histories=train_models()

# -------------------------------------------------
# KPI
# -------------------------------------------------

c1,c2,c3,c4=st.columns(4)

c1.metric("Average CSAT",round(df["csat_score"].mean(),2))
c2.metric("Total Interactions",len(df))

best_model=min(MODEL_SCORES,key=lambda x:MODEL_SCORES[x]["RMSE"])

c3.metric("Best Model",best_model)
c4.metric("Best RMSE",round(MODEL_SCORES[best_model]["RMSE"],3))

# -------------------------------------------------
# SIDEBAR
# -------------------------------------------------

st.sidebar.header("Customer Interaction Details")

item_price = st.sidebar.slider("Item Price",0,100000,500)

handling_time = st.sidebar.slider("Handling Time (minutes)",0,200,10)

response_time = st.sidebar.slider("Response Time (minutes)",0,200,20)

product_category = st.sidebar.selectbox(
"Product Category",
["Electronics","LifeStyle","Fashion","Appliances"]
)

support_channel = st.sidebar.selectbox(
"Support Channel",
["Call","Email","Outcall","Chat"]
)

query_category = st.sidebar.selectbox(
"Query Category",
["Product Queries","Refund","Delivery","Technical"]
)

remark = st.sidebar.selectbox(
"Customer Remark",
["Late delivery","Product damaged","Refund request",
"Wrong item received","Payment issue","Technical problem",
"Poor support response","Excellent service"]
)

# -------------------------------------------------
# TABS
# -------------------------------------------------

tab1,tab2,tab3,tab4,tab5,tab6,tab7,tab8 = st.tabs([
"🎯 CSAT Prediction",
"📊 Model Comparison",
"📈 Business Insights",
"👥 Customer Segmentation",
"🧠 Complaint Clustering",
"📑 AI Report",
"❓ FAQs",
"🚀 Future Recommendations"
])

# -------------------------------------------------
# TAB 1 CSAT
# -------------------------------------------------

with tab1:

    st.subheader("Predict Customer Satisfaction")

    model_name=st.selectbox("Select Model",list(models.keys()))

    if st.button("Predict CSAT Score"):

        prediction=baseline_csats[model_name]

        st.success(f"Predicted CSAT Score: {round(prediction,2)}")

        fig=go.Figure(go.Indicator(
        mode="gauge+number",
        value=prediction,
        title={'text':"Predicted CSAT Score"},
        gauge={'axis':{'range':[1,5]}}
        ))

        st.plotly_chart(fig,use_container_width=True)

        if prediction>=4:
            sentiment="Positive 😊"
            rec="Maintain high quality service."
        elif prediction>=3:
            sentiment="Neutral 😐"
            rec="Improve response speed and support."
        else:
            sentiment="Negative 😡"
            rec="Urgent improvement required."

        st.subheader("Customer Sentiment")
        st.info(sentiment)

        st.subheader("Recommendation")
        st.write(rec)

# -------------------------------------------------
# TAB 2 MODEL COMPARISON
# -------------------------------------------------

with tab2:

    st.write("Compare multiple AI models trained on the same dataset.")

    if st.button("Compare Models"):

        table=pd.DataFrame(MODEL_SCORES).T

        st.subheader("Model Comparison Table")
        st.dataframe(table)

        st.subheader("Model Leaderboard")
        st.dataframe(table.sort_values("RMSE"))

        st.subheader("Training Loss Graphs")

        for name,history in histories.items():
            fig=px.line(y=history.history["loss"],title=name)
            st.plotly_chart(fig,use_container_width=True)

# -------------------------------------------------
# TAB 3 BUSINESS INSIGHTS
# -------------------------------------------------

with tab3:

    st.subheader("1 CSAT Distribution")
    st.plotly_chart(px.histogram(df,x="csat_score"))
    st.write("Shows overall satisfaction levels of customers.")

    st.subheader("2 Handling Time vs CSAT")
    st.plotly_chart(px.scatter(df,x="connected_handling_time",y="csat_score"))
    st.write("Longer handling time usually reduces customer satisfaction.")

    st.subheader("3 Item Price vs CSAT")
    st.plotly_chart(px.scatter(df,x="item_price",y="csat_score"))
    st.write("Helps analyze if high price products cause dissatisfaction.")

    st.subheader("4 CSAT by Product Category")
    if "product_category" in df.columns:
        st.plotly_chart(px.box(df,x="product_category",y="csat_score"))
        st.write("Shows which category has more complaints.")

    st.subheader("5 CSAT by Support Channel")
    if "channel_name" in df.columns:
        st.plotly_chart(px.box(df,x="channel_name",y="csat_score"))
        st.write("Identifies best performing support channel.")

    st.subheader("6 City vs CSAT")
    if "customer_city" in df.columns:
        st.plotly_chart(px.box(df,x="customer_city",y="csat_score"))
        st.write("Helps businesses identify cities with poor customer experience.")

# -------------------------------------------------
# TAB 4 SEGMENTATION
# -------------------------------------------------


with tab4:

    seg=df[["item_price","connected_handling_time","csat_score"]].dropna()

    kmeans=KMeans(n_clusters=3)
    seg["segment"]=kmeans.fit_predict(seg)

    st.subheader("Customer Segments Distribution")
    st.plotly_chart(px.histogram(seg,x="segment"))
    st.write("Shows distribution of customer segments.")

    st.subheader("Item Price vs Segment")
    st.plotly_chart(px.scatter(seg,x="item_price",y="csat_score",color="segment"))
    st.write("Identifies high value vs low value customers.")

    st.subheader("Handling Time vs Segment")
    st.plotly_chart(px.scatter(seg,x="connected_handling_time",y="csat_score",color="segment"))

    st.subheader("Segment vs CSAT")
    st.plotly_chart(px.box(seg,x="segment",y="csat_score"))

    st.subheader("Segment vs Item Price")
    st.plotly_chart(px.box(seg,x="segment",y="item_price"))

    st.subheader("Segment vs Handling Time")
    st.plotly_chart(px.box(seg,x="segment",y="connected_handling_time"))

# -------------------------------------------------
# TAB 5 CLUSTERING
# -------------------------------------------------


with tab5:

    remarks=df["customer_remarks"].dropna()

    tfidf=TfidfVectorizer(stop_words="english")

    X=tfidf.fit_transform(remarks)

    kmeans=KMeans(n_clusters=3)

    clusters=kmeans.fit_predict(X)

    cluster_df=pd.DataFrame({"Remark":remarks,"Cluster":clusters})

    st.subheader("Complaint Cluster Table")

    st.dataframe(cluster_df.head(20))

    st.subheader("Cluster Distribution")

    st.plotly_chart(px.histogram(cluster_df,x="Cluster"))

    st.write(
    "Complaint clustering groups similar issues together. "
    "Businesses can quickly identify major recurring problems "
    "and improve customer support processes."
    )


# -------------------------------------------------
# TAB 6 AI REPORT
# -------------------------------------------------

with tab6:

    st.subheader("Select AI Report Format")

    format_choice=st.selectbox("Format",["TXT","Word","PDF"])

    if st.button("Generate AI Report"):

        report=f"""
PROJECT TITLE
AI Customer Experience Analytics

Deep Learning Models Used
• Deep Neural Network
• LSTM
• GRU
• CNN

Model Performance
{pd.DataFrame(MODEL_SCORES).T}



Business Insights
CSAT distribution shows satisfaction trend.

Customer Segmentation
KMeans clustering identifies customer groups.

Complaint Clustering
NLP clustering groups similar complaints.

Conclusion
AI models help businesses improve customer satisfaction. If you need any other information regarding this project, please do ask me.

Project Deployed By:\n\n
Dharminder Singh Virk\n\n
Technology: Streamlit | Deep Learning | NLP
"""

        if format_choice=="TXT":
            st.download_button("Download Report",report,"report.txt")

        if format_choice=="Word":
            doc=Document()
            doc.add_paragraph(report)
            doc.save("report.docx")
            f=open("report.docx","rb")
            st.download_button("Download Report",f,"report.docx")

        if format_choice=="PDF":
            c=canvas.Canvas("report.pdf")
            text=c.beginText(40,800)
            for line in report.split("\n"):
                text.textLine(line)
            c.drawText(text)
            c.save()
            f=open("report.pdf","rb")
            st.download_button("Download Report",f,"report.pdf")

# -------------------------------------------------
# TAB 7 FAQ
# -------------------------------------------------

with tab7:

    faq={
    "What is CSAT?":"Customer satisfaction score.",
    "Why CSAT important?":"Measures service quality.",
    "How AI predicts CSAT?":"Deep learning models.",
    "What is segmentation?":"Grouping customers.",
    "What is clustering?":"Grouping complaints.",
    "Why model comparison?":"Find best model.",
    "What is RMSE?":"Prediction error metric.",
    "What is R2?":"Model accuracy metric.",
    "Why dashboards useful?":"Visual insights.",
    "How businesses benefit?":"Better decision making."
    }

    for q,a in faq.items():
        with st.expander(q):
            st.write(a)

# -------------------------------------------------
# TAB 8 FUTURE
# -------------------------------------------------

with tab8:

    st.subheader("Future Scope")

    st.write("""
This AI dashboard can be further improved by:

• Using Transformer models
• Using ensemble learning
• Using larger datasets
• Integrating real time customer feedback
• Deploying models in production environment
• Adding real time chatbot analytics
""")

    st.subheader("Potential Models")

    future_models=pd.DataFrame({
    "Model":["Transformer","Random Forest","XGBoost","LightGBM"],
    "Purpose":[
    "Advanced NLP analysis",
    "Better prediction accuracy",
    "Boosting model performance",
    "Large scale dataset handling"
    ]
    })

    st.dataframe(future_models)

st.markdown("---")

st.markdown("""
### Project Deployed By

Dharminder Singh Virk\n\n
Phone Number: +91-8198092901\n\n
Technology: Streamlit | Deep Learning | NLP\n\n
""")
```
